## Refitting and updating parameters

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.svm import SVC
X, y = load_iris(return_X_y=True)
```

charge l'ensemble de données Iris à l'aide de scikit-learn et stocke les données dans les tableaux NumPy X (caractéristiques) et y (étiquettes de classe). Voici ce que fait chaque ligne de code

X, y = load_iris(return_X_y=True): Cette ligne charge l'ensemble de données Iris à l'aide de la fonction load_iris et stocke les caractéristiques (variables indépendantes) dans le tableau X et les étiquettes de classe (variable dépendante) dans le tableau y. L'argument return_X_y=True indique à la fonction de renvoyer directement X et y

```python
clf = SVC()
```

```python
clf.set_params(kernel='linear').fit(X, y)
```

crée un modèle de classification à vecteurs de support (SVM) en utilisant la classe SVC de scikit-learn


clf = SVC(): Cette ligne crée une instance de la classe SVC, qui est un modèle SVM par défaut. Par défaut, le noyau (kernel) est défini sur "rbf" (Radial Basis Function).

clf.set_params(kernel='linear'): Cette ligne utilise la méthode set_params pour définir le noyau du modèle SVM sur "linear" (linéaire). En d'autres termes, cela configure le modèle SVM pour utiliser un noyau linéaire plutôt que le noyau radial de base.

.fit(X, y): Cette ligne entraîne le modèle SVM sur les données X (caractéristiques) et les étiquettes de classe correspondantes y. Le modèle SVM linéaire est maintenant entraîné à apprendre à classer les données en utilisant une séparation linéaire entre les classes.

Après avoir exécuté ces lignes de code, le modèle SVM linéaire clf est entraîné et prêt à être utilisé pour effectuer des prédictions ou pour évaluer sa performance sur de nouvelles données.

```python
clf.predict(X[:5])
```

effectue des prédictions en utilisant le modèle SVM entraîné (clf) sur les cinq premières instances de l'ensemble de données Iris (X[:5]). 

Elle retournera les prédictions pour ces cinq instances en fonction des caractéristiques fournies.

Le résultat sera une liste (ou un tableau NumPy) contenant les étiquettes de classe prédites pour ces cinq instances. 

Chaque élément de la liste représente la classe prédite pour une instance donnée.

Par exemple, si vous obtenez le résultat [0, 0, 0, 1, 1], cela signifie que le modèle SVM prédit que les deux premières instances appartiennent à la classe 0, les deux suivantes appartiennent à la classe 1 et la dernière à la classe 0.

Cela vous permet de voir comment le modèle SVM linéaire classifie les cinq premières instances de l'ensemble de données Iris en utilisant la séparation linéaire apprise lors de l'entraînement.

## Multiclass vs. multilabel fitting

```python
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import LabelBinarizer

X = [[1, 2], [2, 4], [4, 5], [3, 2], [3, 1]]
y = [0, 0, 1, 1, 2]

classif = OneVsRestClassifier(estimator=SVC(random_state=0))
classif.fit(X, y).predict(X)
```

Ce Code utilise la stratégie "One-vs-Rest" (OvR) pour effectuer une classification multiclasse à l'aide d'un modèle SVM (Support Vector Machine). 

Importation des modules nécessaires, y compris SVC pour le modèle SVM, OneVsRestClassifier pour la stratégie OvR, et LabelBinarizer pour la binarisation des étiquettes.

Définition des données X (caractéristiques) et des étiquettes y.

Création d'une instance de OneVsRestClassifier avec un modèle SVC en tant qu'estimateur interne. 
Cela signifie que vous utiliserez plusieurs classificateurs SVM pour effectuer la classification multiclasse en utilisant la stratégie OvR.

Entraînement du modèle classif sur les données X et les étiquettes y à l'aide de la méthode fit.

Prédiction des étiquettes de classe pour les mêmes données X en utilisant la méthode predict.

Le modèle classif est configuré pour effectuer la classification multiclasse en utilisant la stratégie OvR avec des classificateurs SVM internes. La méthode predict(X) retourne les prédictions de classe pour les données X fournies, en utilisant l'ensemble des classificateurs OvR internes pour prendre une décision pour chaque instance.

Les résultats seront une liste (ou un tableau NumPy) de prédictions de classe pour chaque instance dans X. Ces prédictions seront basées sur la stratégie OvR, où chaque classificateur SVM interne est responsable de distinguer une classe particulière des autres.







```python
y = LabelBinarizer().fit_transform(y)
classif.fit(X, y).predict(X)
```

Ici, le classificateur est fit() sur une représentation 2d d'étiquettes binaires de y, en utilisant le LabelBinarizer. Dans ce cas, predict() renvoie un tableau 2d représentant les prédictions multilabels correspondantes.

Notez que les quatrième et cinquième instances renvoient toutes des zéros, ce qui indique qu'elles ne correspondent à aucune des trois étiquettes utilisées pour l'ajustement. Avec les sorties multilabels, il est également possible d'attribuer plusieurs labels à une instance :

```python
from sklearn.preprocessing import MultiLabelBinarizer
y = [[0, 1], [0, 2], [1, 3], [0, 2, 3], [2, 4]]
y = MultiLabelBinarizer().fit_transform(y)
classif.fit(X, y).predict(X)
```

Dans ce cas, le classificateur est adapté à des instances auxquelles sont attribuées plusieurs étiquettes. Le MultiLabelBinarizer est utilisé pour binariser le tableau 2d de labels multiples à ajuster. En conséquence, predict() renvoie un tableau 2d avec plusieurs étiquettes prédites pour chaque instance.

# B. Data normalization

```python
import numpy as np

# Create the matrix X
X = np.array([[1, -1, 2],
              [2, 0, 0],
              [0, 1, -1]])
```

```python
# Print the matrix X
print("Matrix X:")
print(X)
```

```python
# mean of X
mean_X = np.mean(X)

# variance of X
variance_X = np.var(X)


print("Mean of X:", mean_X)
print("Variance of X:", variance_X)
```

```python
from sklearn.preprocessing import scale
# Normalize the matrix X
X_normalized = scale(X)
# Print the normalized matrix X
print("Normalized Matrix X:")
print(X_normalized)
# Compute the mean and variance of the normalized matrix X
mean_normalized_X = np.mean(X_normalized)
variance_normalized_X = np.var(X_normalized)
# Print the mean and variance of the normalized matrix X
print("Mean of Normalized X:", mean_normalized_X)
print("Variance of Normalized X:", variance_normalized_X)
```

# Data visualization

```python
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = iris.data
target = iris.target
target_names = iris.target_names

# Create a DataFrame for easier visualization
df = pd.DataFrame(data, columns=iris.feature_names)
df['target'] = target


print(df.head())

# Gliste de tuples (combinations) où chaque tuple représente une paire 
#de variables de caractéristiques. Ces combinaisons seront utilisées pour 
#créer des graphiques de dispersion de différentes paires de caractéristiques.

combinations = [('sepal length (cm)', 'sepal width (cm)'),
                ('sepal length (cm)', 'petal length (cm)'),
                ('sepal length (cm)', 'petal width (cm)'),
                ('sepal width (cm)', 'petal length (cm)'),
                ('sepal width (cm)', 'petal width (cm)'),
                ('petal length (cm)', 'petal width (cm)')]

# Create subplots for each combination
plt.figure(figsize=(15, 10))
for i, (x_var, y_var) in enumerate(combinations, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(data=df, x=x_var, y=y_var, hue='target', palette='viridis')
    plt.title(f'Combination: {x_var} vs {y_var}')
    plt.xlabel(x_var)
    plt.ylabel(y_var)

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Compute the correlations between each pair of variables
correlations = np.corrcoef(data, rowvar=False)

# Print the correlation matrix
print("Correlation Matrix:")
print(correlations)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names

# Create a scatter plot for all possible combinations of variables
n_features = X.shape[1]
fig, axes = plt.subplots(n_features, n_features, figsize=(12, 12))

for i in range(n_features):
    for j in range(n_features):
        if i != j:
            ax = axes[i, j]
            ax.scatter(X[:, i], X[:, j], c=y, cmap=plt.cm.Set1, edgecolor='k')
            ax.set_xlabel(feature_names[i])
            ax.set_ylabel(feature_names[j])
            ax.set_xticks(())
            ax.set_yticks(())

# Add a legend
handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=plt.cm.Set1(0), markersize=10, label=target_names[0]),
           plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=plt.cm.Set1(1), markersize=10, label=target_names[1]),
           plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=plt.cm.Set1(2), markersize=10, label=target_names[2])]
plt.legend(handles=handles, title='Classes', loc='upper left')

plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import datasets

# Load the Iris dataset
iris = datasets.load_iris()
data = iris.data

# Perform PCA
pca = PCA(n_components=2)  # Specify the number of components (2 for 2D visualization)
IrisPCA = pca.fit(iris.data).transform(iris.data)

# Perform LDA
lda = LinearDiscriminantAnalysis(n_components=2)  # Specify the number of components (2 for 2D visualization)
IrisLDA = lda.fit(iris.data, iris.target).transform(iris.data)

# Print the first few rows of the transformed data
print("PCA Transformed Data (IrisPCA):")
print(IrisPCA[:5])  # Print the first 5 rows of PCA results

print("\nLDA Transformed Data (IrisLDA):")
print(IrisLDA[:5])  # Print the first 5 rows of LDA results
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Apply PCA and save the results in IrisPCA
pca = PCA(n_components=2)
IrisPCA = pca.fit_transform(X)

# Apply LDA and save the results in IrisLDA
lda = LDA(n_components=2)
IrisLDA = lda.fit_transform(X, y)
print(IrisLDA[:5])
print(IrisPCA[:5])
```

```python
# Plot the PCA results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(IrisPCA[:, 0], IrisPCA[:, 1], c=y)
plt.title('PCA Projection')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.xlim(IrisPCA[:, 0].min() - 1, IrisPCA[:, 0].max() + 1)
plt.ylim(IrisPCA[:, 1].min() - 1, IrisPCA[:, 1].max() + 1)

# Plot the LDA results
plt.subplot(1, 2, 2)
plt.scatter(IrisLDA[:, 0], IrisLDA[:, 1], c=y)
plt.title('LDA Projection')
plt.xlabel('Linear Discriminant 1')
plt.ylabel('Linear Discriminant 2')
plt.xlim(IrisLDA[:, 0].min() - 1, IrisLDA[:, 0].max() + 1)
plt.ylim(IrisLDA[:, 1].min() - 1, IrisLDA[:, 1].max() + 1)

plt.tight_layout()
plt.show()
```
